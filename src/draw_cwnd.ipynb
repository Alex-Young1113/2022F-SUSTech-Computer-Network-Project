```python
import matplotlib.pyplot as plt
time = []
cwnd = []
with open('../log/peer2.log', 'r') as f:
    lines = f.readlines()
    for l in lines:
        if 'cwnd' in l:
            time.append(float(l.split(' ')[-2]))
            cwnd.append(float(l.split(' ')[-1]))
plt.plot(time, cwnd)
plt.show()
```
